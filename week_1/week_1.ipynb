## Глоссарий

__Базовый логический элемент__ - цифровой элемент, выполняющий базовую логическую операцию(например, инверсия, дизъюнкция, конъюнкция)

__Детерминированный конечный автомат__ - такой конечный автомат, в котором в каждом состоянии по каждому входному символу определено не более одного перехода

__Диагностическая последовательность__ - входная  последовательность,  по  реакции  на которую можно различить все состояния автомата

__Идентификатор  состояния__ - конечное  множество  входных  последовательностей, отличающих данное состояние от всех остальных состояний автомата

__Инициальный автомат__ - это  такой  автомат,  в  котором
- выделено  особое  состояние, называемое начальным;
- работа  всегда  начинается  с  этого  состояния;
- есть  надежный сигнал сброса, подача которого возвращает автомат в начальное состояние

__Комбинационная логическая схема__ - это  устройство,  не  имеющее памяти,  то  есть выходные  сигналы  такого  устройства  определяются  только  текущими  значениями входных сигналов в предыдущий момент времени

__Конечный автомат__ - пятерка $A=(S,I,O,\psi,\phi)$, где 
- $S$ - конечное множество состояний, 
- $I$ - конечное  множество  входных  символов, 
- $O$ - конечное  множество  выходных  символов, 
- $\psi : S \times I\to S$ - функция переходов,  
- $\phi : S \times I \to O$ - функция выходов

__Конечный  автомат__ (определение  с  учетом  возможной  частичности) - шестерка $A=(S,I,O,D_A,\psi,\phi)$, где 
- $S$ - конечное  множество  состояний, 
- $I$ - конечное  множество входных  символов, 
- $O$ - конечное  множество  выходных  символов, 
- $D_A \subseteq S \times I$ - область определения функции переходов и функции выходов автомата (т.е. $D_A$ - множество  пар  (состояние,  входной  символ),  на  которых  поведение  автомата определено),
- $\psi : D_A \to S$ - функция переходов,
- $\phi : D_A \to O$ - функция выходов

__Конечный  автомат__ (определение  с  учетом  возможного  недетерминизма) - четверка $A=(S,I,O,h)$, где 
- $S$ - конечное множество состояний, 
- $I$ - конечное множество входных символов, 
- $O$ - конечное  множество  выходных  символов, 
- $h \subseteq S \times I \times S \times O$ - отношение переходов-выходов, определяющее всевозможные переходы в автомате

__Минимальный__ (или приведенный) __автомат__ - такой конечный автомат,который имеет минимально возможное число состояний для заданного поведения

__Множество  достижимости  автомата__ - конечное множество    входных последовательностей, по которым достигается каждое состояние из начального состояния

__Множество различимости автомата__ - объединение идентификаторов  всех  состояний автомата

__Модель неисправности__ - это тройка $<A, \sim, R>$, где 
- $A$ - эталонный автомат,
- $R$ - множество проверяемых автоматов с тем же входным алфавитом, что и у автомата $A$, и числом состояний не  более, чем у $A$,
- $\sim$ - отношение соответствия (комформности) между эталонным и проверяемым автоматами

__Начальное состояние автомата__ - такое состояние, что
- с которого всегда начинается работа автомата; 
- есть надежный сигнал сброса, 
- подача которого возвращает автомат в начальное состояние

__Недетерминированный конечный автомат__ - такой  конечный  автомат, в  котором в каком-то по крайней мере одном состоянии по  одному  или  нескольким  входным  воздействиям определено более одного перехода

__Полностью определенный конечный автомат__ - такой конечный  автомат,  в  котором определены  переходы  из  каждого  состояния  по  каждому  входному  воздействию,  т.е. $D_A=S \times I$

__Полный  проверяющий  тест  относительно  модели  неисправности  $<A,   \sim, R>$__ - конечное  множество  конечных  входных  последовательностей,  по  реакциям  на  которые можно отличить всякую неконформную реализацию из области неисправности

__Последовательностная схема__ - это схема, значения выходных сигналов которой зависят не только от значений входных сигналов в данный момент, но и от внутренних состояний схемы в предыдущий момент времени

__Приведенный__ (или минимальный) __автомат__ - такой конечный автомат,который имеет минимально возможное число состояний для заданного поведения

__Разбиение $P_k$__($k$-эквивалентное  разбиение  автомата) - такое  разбиение  множества состояний  автомата,  что в  каждом  классе  разбиения  содержатся  состояния,  в  которых автомат   производит   одинаковые   выходные   реакции   на   любую   входную последовательность длины $k$

__Разбиение $P$__ (эквивалентное разбиение автомата) - такое разбиение множества состояний автомата на  классы  эквивалентности,  что в  каждом  классе  разбиения  содержатся состояния,  в  которых  автомат  производит  одинаковые  выходные  реакции  на  любую входную последовательность (любой длины)

__Различающая  последовательность__ - входная  последовательность,  по  реакции  на которую различаются два состояния автомата

__Различимые  инициальные  автоматы__ - такие  автоматы, у  которых различимы начальные состояния

__Различимые  состояния  автомата__ - такие  состояния,  что найдется  такая  входная последовательность, на которую выходные реакции автомата в данных состояниях будут различаться

__Регистр__ - запоминающее  устройство,  предназначенное  для  временного  хранения информации

__Регистр сдвига__ - это регистр,  обеспечивающий  сдвиг  влево  или вправо  всех  разрядов одновременно на одинаковое число позиций

__Связный  автомат__ - автомат,  в  котором  все  состояния  достижимы  из  начального состояния

__Синхронные триггеры__ - это триггеры, которые реагируют на информационные сигналы только при наличии соответствующего сигнала на входе синхронизации

__Счётчик__ - это  устройство,  на  выходах  которого  получается  двоичный  или  двоично-десятичный код, определяемый числом входных импульсов

__Триггер__ - дискретное  устройство,  которое  способно  длительно  находиться  в  одном  из двух  устойчивых  состояний  и  изменять  свое  состояние  под  воздействием  внешних сигналов

__Цифровые сигналы__ - это дискретные сигналы, которые имеют только два допустимых значения ($0$ и $1$)

__Частичный конечный автомат__ - такой  конечный  автомат,  в  котором в  каком-то  по крайней  мере  одном  состоянии  по  одному  или  нескольким  входным  воздействиям  не определено ни одного перехода, т.е. $DA \subset S \times I$

__Эквивалентные состояния автомата__ - такие состояния, что выходные реакции автомата в данных состояниях на любую входную последовательность совпадаю

__Эквивалентные инициальные автоматы__ - такие  автоматы, у  которых  эквивалентны начальные состояния

### Конечный автомат

__Конечный автомат__ - пятерка $A=(S,I,O,\psi,\phi)$, где 
- $S$ - конечное множество состояний (States), 
- $I$ - конечное  множество  входных  символов (Input), 
- $O$ - конечное  множество  выходных  символов (Output), 
- $\psi : S \times I \to S$ - функция переходов,  
- $\phi : S \times I \to O$ - функция выходов

#### Конечный автомат: функция переходов

$$\psi : S \times I\to S$$

Функция переходов 
- определяет *переход* из состояния по входному символу в следующее состояние
- ставит в соответствие паре *(состояние, входной символ)* некоторое новое состояние

#### Конечный автомат: функция переходов

$$\phi : S \times I \to O$$

Функция переходов 
- определяет *выходную реакцию* в текущем состоянии под действием входного символа
- ставит в соответствие паре *(состояние, входной символ)* некоторый выходной символ

### Сумматор

![1_2_2.png](1_2_2.png)

- Последовательный сумматор имеет два входа $I_1$ и $I_2$ и один выход $O$
- На входы сумматора подаются слагаемые, представленные в двоичной системе счисления: алфавиты входных каналов $I_1$ и $I_2$ - это $\{0, 1\}$ 
- Алфавит входного канала $O$ - это также $\{0, 1\}$
- Слагаемые подаются последовательно младшими разрядами вперед 
- На выходе выдается их сумма в двоичной системе счисления
- Состояния сумматора - это $0$ и $1$: то, что хранится в памяти (перенос разряда)

### Функция переходов для автомата сумматора

#### Функция переходов в состоянии 0 

При суммировании в двоичной системе счисления разряд переноса возникает только в случае, если складывались две единицы, поэтому функция переходов из состояния $0$:
- под действием входных символов $00$, $01$ и $10$ равна $0$ 
- под действием входного воздействия $11$ равна $1$

$$\psi(0,00)=0$$
$$\psi(0,01)=0$$
$$\psi(0,10)=0$$
$$\psi(0,11)=1$$

#### Функция переходов в состоянии 1 

При суммировании в двоичной системе счисления сброс разряда переноса (переход в состояние $0$) возникает только если складывались два нуля, поэтому функция переходов из состояния $1$:
- под действием входного воздействия $00$ равна $0$
- под действием входных воздействий $01$, $10$, $11$ равна $1$

$$\psi(1,00)=1$$
$$\psi(1,01)=1$$
$$\psi(1,10)=1$$
$$\psi(1,11)=1$$

### Построение выходной последовательность автомата по заданной входной последовательности

Пусть дан конечный автомат $A=(S,I,O,\psi,\phi)$, где 
- $S$ - конечное множество состояний, 
- $I$ - конечное  множество  входных  символов, 
- $O$ - конечное  множество  выходных  символов, 
- $\psi : S \times I \to S$ - функция переходов,  
- $\phi : S \times I \to O$ - функция выходов

Построим выходную последовательность автомата $A$, если во время подачи входной последовательности $\alpha = i_1 i_2 ... i_k$ находился в некотором состоянии $s$. 

Для  того, чтобы найти  выходную  последовательность,  нужно промоделировать работу автомата $A$ на данной входной последовательности, начиная с состояния $s$.

С  помощью  функции  переходов $\psi$ будем  определять  следующее состояние, а с помощью функции $\phi$ - выходную реакцию автомата. Всего нужно  будет  произвести $k$ таких  вычислений, где $k$ - длина последовательности $\alpha$.

Пусть $s_1 = s$ - состояние, с которого начинаем вычисления. Определим первый выходной символ $o_1$ с помощью функции $\phi$: $o_1 = \phi(s_1, i_1)$. Для вычисления следующего выходного символа нам нужно знать, в какое  состояние  перешел  автомат  из состояния $s_1$ по  входному  символу $i_1$. Определим это с помощью функции $\psi$: $s_2 = \psi(s_1, i_1)$. Значит, входной символ $i_2$ нужно подать в состоянии $s_2$. 

Последний выходной символ $o_k = \phi(s_k, i_k)$, где $s_k$ вычислено по состоянию $s_{k-1}$ и входному символу $i_{k-1}$.

Последовательность $\beta = o_1 o_2 ... o_k$,  построенная  описанным  способом, будет являться выходной последовательностью автомата $A$ в состоянии $s$ на входную последовательность $\alpha = i_1 i_2 ... i_k$. 

### Пример построения конечного автомата для подсчета слов с определенными характеристиками в заданном тексте

Опишем конечным автоматом подсчет слов в английском тексте, начинающихся с *un* и оканчивающихся на *d* (understand, united и т.д.). Получается, что на вход автомата поступает английский текст, составленный из 26 букв алфавита и пропусков. На выход  будет выдаваться **yes**, если встретилось очередное слово, начинающееся на **un** и оканчивающееся на **d**, а **no** - если очередное слово не удовлетворяетуказанным требованиям, либо еще не является словом (будем считать, что после слова должен следовать пробел). 

Для простоты пропуски букв обозначим буквой $\pi$ , а все другие буквы, кроме *d*, *n* и *u* буквой $\lambda$.

Входной алфавит автомата $I = \{d, n, u, \pi, \lambda\}$.

Выходной  алфавит  автомата $O = \{yes, no\}$, где **yes** означает «считать  данное слово», **no** - «не считать». 

$S = \{$new word (новое слово), waiting for new word (ждать нового слова), appearing u (появление u), appearing u-n (появление u-n), appearing u-n-d (появление u-n-d)$\}$ 

![1_2_18.png](1_2_18.png)

Рисунок 1. Автомат по подсчету слов в английском тексте, начинающихся с *un* и оканчивающихся на *d*

При  входе $\pi$ происходит переход в состояние «новое  слово»,  независимо  от  текущего состояния,  поскольку  пробел означает,  что  очередное  слово  закончено. При  этом  при текущем  состоянии «появление u-n-d» и  входе $\pi$ появляется  выход **yes** (считать), поскольку  мы  дождались  слова,  удовлетворяющего  нашим  требованиям.  Во  всех остальных  случаях - выход **no** (не  считать),  потому  что  слово  не  соответствует оговоренным требованиям. 

При  текущем  состоянии «новое  слово» и  входе *u* переходим  в  состояние «появление u» (слово началось), а при входе *d*, *n* или $\lambda$ - в состояние «ждать нового слова» (слово началось не с буквы u). 

При  текущем  состоянии «появление u» и  входе *n* будет  переход  в состояние «появление u-n» (слово  и  началось,  и  продолжилось), а при входе *d*, *u* или $\lambda$ - в состояние «ждать нового слова». 

При текущем состоянии «появление u-n» или «появление u-n-d» при входе *d* следующее  состояние - «появление u-n-d», а  при  входе *u*, *n* или $\lambda$ следующее  состояние - «появление u-n». 

Состояние «ждать нового слова» при входе, отличном от $\pi$, остается неизменным.

#### Построение выходной последовательности по заданной входной последовательности для данного автомата

Пусть  на  вход  автомата подается  последовательность  «united states».  Во входном алфавите $I = \{d, n, u, \pi, \lambda\}$, где пропуски букв обозначены буквой $\pi$, а все другие буквы, кроме *d*, *n* и *u* - буквой $\lambda$, имеем входную последовательность $\alpha = u n \lambda \lambda \lambda d \pi \lambda \lambda \lambda \lambda \lambda \lambda$.

Построим выходную последовательность $\beta$, стартуя из состояния «новое слово». 

По входному символу *u* автомат перейдет в состояние «появление u», на выходе - **no**. 

Из состояния «появление u» автомат по входному символу *n* перейдет в состояние «появление u-n», на выходе - **no**.

Из  состояния «появление u-n» автомат  по  входному  символу $\lambda$ останется  в состоянии «появление u-n», на выходе - **no** (по двум следующим входным символам $\lambda$ будет то же самое).

Из  состояния «появление u-n» автомат  по  входному  символу dостанется  в состоянии «появление u-n-d», на выходе - **no**.

Из  состояния «появление u-n-d» автомат  по  входному  символу $\pi$ перейдет  в состояние «новое слово», на выходе - **yes**. 

Из состояния «новое слово» автомат по входному символу $\lambda$ перейдет в состояние «ждать нового слова», на выходе - **no**.

Из состояния «ждать нового слова» автомат по входному символу $\lambda$ останется в состоянии «ждать  нового  слова»,  на  выходе - **no** (по  четырем  следующим  входным символам $\lambda$ будет то же самое).

Итак,  на  входную  последовательность $\alpha = u n \lambda \lambda \lambda d \pi \lambda \lambda \lambda \lambda \lambda \lambda$.\ автомат  выдаст выходную последовательность $\beta = $ **no** **no** **no** **no** **no** **no** **yes** **no** **no** **no** **no** **no** **no**.

### Конечный автомат

#### Определение  с  учетом  возможной  частичности

__Конечный  автомат__ - шестерка $A=(S,I,O,D_A,\psi,\phi)$, где 
- $S$ - конечное  множество  состояний, 
- $I$ - конечное  множество входных  символов, 
- $O$ - конечное  множество  выходных  символов, 
- $D_A \subseteq S \times I$ - область определения функции переходов и функции выходов автомата (т.е. $D_A$ - множество  пар  (состояние,  входной  символ),  на  которых  поведение  автомата определено),
- $\psi : D_A \to S$ -  функция переходов,
- $\phi : D_A \to O$ - функция выходов

#### Определение  с  учетом  возможного  недетерминизма

__Конечный  автомат__ - четверка $A=(S,I,O,h)$, где 
- $S$ - конечное множество состояний, 
- $I$ - конечное множество входных символов, 
- $O$ - конечное  множество  выходных  символов, 
- $h \subseteq S \times I \times S \times O$ - отношение переходов-выходов, определяющее всевозможные переходы в автомате

Элементами множества $h$ являются четверки вида $(s, i, s', o)$, называемые *переходами*

Автомат может перейти из состояния $s \in S$ под действием входного символа $i \in I$ в состояние $s' \in S$ с выдачей выходного символа $o \in O$, если четверка $(s, i, s', o)$ содержится в $h$

### Конечные автоматы

__Полностью определенные__ - определены переходы из каждого состояния по каждому входному воздействию ($D_A = S \times I$ - множество $D_A$ совпадает с декартовым произведением множества состояний и множества входных воздействий)

__Частичные__ - в каком-то по крайней мере одном состоянии по одному или нескольким входным воздействиям не определены переходы ($D_A \subset S \times I$ - множество $D_A$ есть собственное подмножество декартова произведения множества состояний и множества входных воздействий)

__Детерминированные__ - в каждом состоянии по каждому входному символу определено не более одного перехода

__Недетерминированные__ - в каком-то по крайней мере одном состоянии по одному или нескольким входным воздействиям определено более одного перехода

### Возникновение частичности
    
Основная причина возникновения частичности в модели - это не полностью заданное неформальное описание поведения устройства или системы

Обычно спецификации протоколов частичны

### Возникновение недетерминизма
    
- доопределение частичности различными способами

- описание функционирования устройства/системы при различных (внешних) условиях

- описание возможных неисправностей устройства/системы
